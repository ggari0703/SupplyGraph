```python
import pandas as pd
import numpy as np

# Load your data
production_data = pd.read_csv("Raw Dataset/Homogenoeus/Temporal Data/Unit/Production .csv")
sales_order_data = pd.read_csv("Raw Dataset/Homogenoeus/Temporal Data/Unit/Sales Order.csv")

# Extract item columns (all columns except the date column)
date_column = 'Date'
item_columns = [col for col in production_data.columns if col != date_column]

# Generate random prices for each item
np.random.seed(42)  # for reproducibility
prices = np.random.uniform(10, 100, size=len(item_columns)).round(2)

# Create a DataFrame with item names and their respective prices
prices_df = pd.DataFrame({'Item': item_columns, 'Price': prices})

# Save to CSV
prices_df.to_csv('Raw Dataset/Item_Prices.csv', index=False)

# Display the resulting DataFrame
prices_df
```
